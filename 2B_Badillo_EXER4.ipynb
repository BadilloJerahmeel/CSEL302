```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

```

```python
data_path = "Lesson_4_student_grades_dataset.csv"
dataset = pd.read_csv(data_path)
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

```python
Hours_of_Study = data.drop('Grades', axis = 1)

Grades = data[['Grades']]
```

```python
Hours_of_Study_train, Hours_of_Study_test, Grades_train, Grades_test = train_test_split(Hours_of_Study, Grades, test_size=0.2, random_state=2)

model = LinearRegression()
model.fit(Hours_of_Study_train, Grades_train)
```

```python
plt.scatter(data['Hours_of_Study'],data['Grades'])
plt.xlabel('Hours_of_Study')
plt.ylabel('Grades')
plt.title('Scatter of plot data')
plt.show()
```

```python
plt.scatter(data['Hours_of_Study'], data ['Grades'])
plt.plot(Hours_of_Study,model.predict(Hours_of_Study), color = 'black')
plt.xlabel('Hours_of_Study')
plt.ylabel('Grdes')
plt.title('Linear Regression')
plt.show()
```

```python
new_data = np.array ([[2,70,85,1,50]])

prediction = model.predict(new_data)

print(f"Predict Grade:{prediction[0]}")
```
